16. splitting string into multiple columns

```python
import pandas as pd
df= pd.DataFrame({'name':['Ahmed Raza','ali','ahmed','khan','azmat','akram','saeed','alizeh','aslam','sabar'],
                  'location':['karachi, Pakistan','lahore, Pakistan',
                              'islamabad, Pakistan','quetta, Pakistan','lahore, Pakistan','islamabad, Pakistan','quetta, Pakistan','lahore, Pakistan','islamabad, Pakistan',
                              'quetta, Pakistan']  }
                 )
```

```python
df
```

```python
df.name.str.split(' ',expand=True)
```

```python
df[['first_name','last_name']]=df.name.str.split(' ',expand=True)
df
```

```python
df[['location_1','location_2']]=df.location.str.split(',',expand=True)
df
```
