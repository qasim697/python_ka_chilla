## PLOTTING IN PYTHON

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
kashti = sns.load_dataset('titanic')
kashti.head()
```

```python
kashti.info()
```

```python
kashti.isnull().sum()
```

```python
kashti["survived"].value_counts().plot(kind='bar')
```

```python
sns.countplot(x='survived', data=kashti, hue='sex')
```

```python
# scatter plot
plt.scatter(x=kashti['age'], y=kashti['fare'], color='purple', alpha=0.5)
```

```python
sns.scatterplot(x='age', y='fare', hue='sex', data=kashti, palette='viridis')
```
