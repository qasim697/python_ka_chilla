## PLOTTING IN PYTHON

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
kashti = sns.load_dataset('titanic')
kashti.head()
```

```python
kashti.info()
```

```python
kashti.isnull().sum()
```

```python
kashti["survived"].value_counts().plot(kind='bar')
```

```python
sns.countplot(x='survived', data=kashti, hue='sex')
```

```python
# scatter plot
plt.scatter(x=kashti['age'], y=kashti['fare'], color='purple', alpha=0.5)
```

```python
sns.scatterplot(x='age', y='fare', hue='sex', data=kashti, palette='plasma')
```

```python
sns.scatterplot(x='ag)e', y='fare', hue='pclass', data=kashti, palette='inferno'
```

```python
sns.barplot(x='sex', y='fare', hue='pclass', data=kashti, palette='viridis')
```

```python
sns.boxplot(x='sex', y='fare', hue='pclass', data=kashti, palette='inferno')
```

```python
sns.pairplot(kashti)
```

```python
kashti.hist(bins=7, figsize=(20,20), color='g', grid=False)
```

```python
sns.stripplot(x='pclass', y='fare', data=kashti, palette='viridis')
```
