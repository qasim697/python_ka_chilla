# Exploratory Data analysis
EDA is a method used to analyze and summarize datasets. in  python, you can use pandas library to perform EDA on a dataset.

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= sns.load_dataset('titanic')
```

```python
# visualize the nan values in heatmap
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
# deck is mostly null values so we can drop it
df.drop('deck',axis=1,inplace=True)
```

```python
df.shape
```

```python
#visualize and explore the data
sns.set_style('whitegrid')
sns.countplot(x='survived',data=df)
```

```python
sns.barplot(data=df,x='survived',y='age',hue='who')
```

```python
sns.barplot(data=df,x='survived',y='fare',hue='who')
```

```python
sns.countplot(data=df,x='pclass',hue='who')
```

```python
#  visualize the age distribution
sns.histplot(df['age'].dropna(),bins=30,kde=True)
```

```python
sns.histplot(df['age'],bins=5)
```

```python
sns.histplot(df['age'],bins=30)
```

```python

# Calculate Q1, Q2, Q3, and IQR for the 'age' column
Q1 = df['age'].quantile(0.25)
Q2 = df['age'].median()
Q3 = df['age'].quantile(0.75)
IQR = Q3 - Q1

print(f"Q1: {Q1}")
print(f"Q2 (Median): {Q2}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")
```

### RESOLVE OUTLIER

```python
plt.figure(figsize=(20, 10))
sns.scatterplot(data=df,x='age',y='fare',hue='who')
plt.ylim(0,300)
```

```python
sns.pairplot(df, hue='survived' )
```

```python
df.groupby('who')['age'].mean()
```

```python
df.groupby(['pclass','sex'])['survived'].mean().unstack().plot(kind='bar',stacked=True)
```

```python
df.groupby(['pclass','who'])['survived'].mean().unstack().plot(kind='bar',stacked=True)
```

```python
# swarm plot
sns.swarmplot(data=df,x='survived',y='age',hue='who')
```

```python
sns.catplot(data=df,x='pclass',y='fare',hue='embarked',kind='point')
```

```python
#relation oftwo numerical columns, lm plot
sns.lmplot(data=df,x='fare',y='age',hue='survived')
```

```python
# Using conditional arguments in sns.lmplot with fare < 300
filtered_df = df.query('fare < 300')
sns.lmplot(data=filtered_df, x='age', y='fare', hue='survived')
```

```python
# how to add a column and make age groups
sns.histplot(df['age'],bins=30)
```

```python
#grouping the age
df['age_group'] = pd.cut(df['age'],bins=[0,18,30,50,80],labels=['child','adult','middle_age','old'])
```

```python
df.head()
```

```python
df.groupby('age_group')['survived'].mean().plot(kind='bar')
```

```python
df.groupby('age_group')['survived'].value_counts()
```

```python
sns.pairplot(df,hue='survived', vars=['age','fare', 'pclass'])
```
